```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
plt.style.use('seaborn-poster')
```

```python
files = ["exp5.csv", "exp10.csv", "exp15.csv"]

df2 = pd.read_csv(files[0], names=["nodos", "conflictos"]).groupby("nodos").mean()
df2 = df2.rename(columns={'conflictos': 'dificultad=5'})
df5 = pd.read_csv(files[1], names=["nodos", "conflictos"]).groupby("nodos").mean()
df2["dificultad=10"] = df5["conflictos"]
df10 = pd.read_csv(files[2], names=["nodos", "conflictos"]).groupby("nodos").mean()
df2["dificultad=15"] = df10["conflictos"]

df2.plot.bar(logy=True).set_ylabel("conflictos")

plt.show()
```
